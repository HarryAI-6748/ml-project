```python
import pandas as pd
import numpy as np
hlth_df = pd.DataFrame(hlth_data)
hlth_df
```

Converting age column (i.e object) to numeric
dup_df['age'] = dup_df['age'].astype(int) will not work as the data has null values, which raises coerce error. To convert this type of data, we need to use "pd.to_numeric(df['age'], errors='coerce')"

```python
hlth_df['age'] = pd.to_numeric(hlth_df['age'], errors = 'coerce')
```

Checking the data type after conversion

Converting weight column (i.e object) to numeric

```python
hlth_df['weight'] = pd.to_numeric(hlth_df['weight'], errors = 'coerce')
```

Checking the data type after conversion

```python
hlth_df.dtypes
```

Checking how many NEW missing values were created after conversion

```python
x = hlth_df.isnull().sum().sum() - hlth_df_original.isnull().sum().sum()
print(f"The number of NEW missing values is: {x}")
```

Converting insurance_provider to category type using astype('category')

```python
hlth_df['insurance_provider'] = hlth_df['insurance_provider'] = hlth_df['insurance_provider'].astype('category')
```

Checking the data type after conversion

```python
hlth_df.dtypes
```

Handling Missing Values in age column using fillna method



```python
hlth_df['age'] = hlth_df['age'].fillna(hlth_df['age'].median())
hlth_df['age']
```

Handling Missing Values in weight column using fillna method


```python
hlth_df['weight'] = hlth_df['age'].fillna(hlth_df['weight'].median())
hlth_df['weight']
```

Handling Missing Values in blood_pressure column using fillna method



```python
hlth_df['blood_pressure'] = hlth_df['age'].fillna(hlth_df['blood_pressure'].median())
hlth_df['blood_pressure']
```

Handling Missing Values in medication column using fillna method


```python
hlth_df['medication'] = hlth_df['medication'].fillna(hlth_df['medication'].mode()[0])
hlth_df['medication']
```

```python
hlth_df
```

Handling Missing Values in insurance_provider column using fillna method


```python
hlth_df['insurance_provider'] = hlth_df['insurance_provider'].cat.add_categories(['Unknown']).fillna('Unknown')
hlth_df['insurance_provider']
```

Verifying number missing values

```python
print(hlth_df.isnull().sum().sum())
```

Getting the duplicate rows using duplicated() method using argument "keep = false" which will mark all the duplicate rows as "True"

Identifying duplicates based on patient_id only using df.duplicated(subset=['patient_id'])
