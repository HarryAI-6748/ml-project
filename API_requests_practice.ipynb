```python
from together import Together
client = Together(api_key = Api_key)
city = input(" Pls enter the name of city: ")
prompt = f"For the given city, provide the latitude and longitude values in the form of json. City: {city}. Only provide me the key for latitude and longitude. Add another key 'current_weather'=True. Do not print any extra detail except the json object"
response = client.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    stream=False,
)
content = response.choices[0].prompt.content
print(content)
```
