```python
import json
with open("inventory.json", "r") as file:#("Converted JSON string into Python list")
  inventory = json.load(file)
  books = 0
  for items in inventory:#("Used a For loop to count the total number of objects")
    books += 1
  print("Total number of books in inventory is:",books, end="\n\n")
new_book = {"title": "Atomic Habits", "author": "James Clear", "price": 14.99, "in_stock": True}
inventory.append(new_book)#("Added a new object to the list")
#print(inventory)
with open("inventory.json", "w") as updated_file:#("Used With open command to open inventory.json file in write mode")
  json.dump(inventory, updated_file, indent=4)#("Used json.dump function to re-write the new content onto the existing file")
for items in inventory:#("Used the for loop to print the items in desired format")
  print(f"Title: {items["title"]} | Author: {items["author"]} | Price: ${items["price"]}")
```
