```python
import numpy as np

data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])

mean_temp = data.mean(axis = 1)
mean_temo_1 = mean_temp[:, np.newaxis]
print("Mean temperature per station")
print(mean_temo_1, end ="\n\n")
high_temp = (data > 28.0)
print("Temperature readings above 28.0°C")
print(data[high_temp], end="\n\n")
x = np.max(data)
y = np.min(data)
z = x - y
normalised = (data - y) / z
normalized = np.round(normalised, 2)
print("Normalized data")
print(normalized)


```
