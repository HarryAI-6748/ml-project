```python
import pandas as pd
import numpy as np

np.random.seed(42)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [12000, 15000, 13500, 17000, 16000, 19000,
         18500, 21000, 20000, 22500, 24000, 26000]

subjects = ['Math', 'Science', 'English', 'History', 'Art']
scores   = [78, 85, 72, 90, 65]

study_hours = np.random.uniform(1, 10, 100)
exam_scores = study_hours * 8 + np.random.normal(0, 5, 100)

product_sales = np.random.normal(loc=50, scale=10, size=200)
```

Using matplotlib, plot monthly sales over the 12-month period. Use months on the x-axis and sales on the y-axis. Customize the line color, add markers, and include a title and axis labels.

```python
import matplotlib.pyplot as plt

plt.plot(months, sales,
color='red',
linestyle='--',
linewidth=2,
marker='o',
label='Sales')
plt.title("Monthly Sales Trend '2025'")
plt.xlabel("Sale Figures")
plt.ylabel("Month")
plt.legend()
plt.show()
```

Using matplotlib, create a bar plot comparing scores across five subjects. Add a title and axis labels. Make sure each bar is clearly readable.

```python
plt.bar(subjects, scores)
plt.title("Subject Score Comparison")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

```python
colors = ['steelblue', 'coral', 'mediumseagreen', 'goldenrod', 'skyblue']
plt.bar(subjects,
        scores,
        color = colors,
        width=0.6,
        edgecolor='black',
        label='Scores')
plt.legend()
plt.title("Subject Score Comparison")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

```python
x = np.arange(len(subjects)) # [0, 1, 2, 3]
width = 0.60 # half-width of each bar
fig, ax = plt.subplots()
bar = ax.bar(x - width/2, scores, width, label='Scores', color=colors)
ax.bar_label(bar, fmt='%d', padding=3)
ax.set_xticks(x)
ax.set_xticklabels(subjects)
ax.set_title("Subject Score Comparison")
ax.set_xlabel("Subject")
ax.set_ylabel("Scores")
ax.legend()
plt.show()
```

```python
print("Its just a example from the notes", '\n')
import numpy as np
import matplotlib.pyplot as plt
categories = ['North', 'South', 'East', 'West']
q1_sales = [45000, 62000, 38000, 71000]
q2_sales = [51000, 59000, 44000, 68000]
x = np.arange(len(categories)) # [0, 1, 2, 3]
width = 0.35 # half-width of each bar
fig, ax = plt.subplots()
bar_1 = ax.bar(x - width/2, q1_sales, width, label='Q1', color='steelblue')
bar_2 = ax.bar(x + width/2, q2_sales, width, label='Q2', color='coral')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.bar_label(bar_1, fmt='%d', padding=3)
ax.bar_label(bar_2, fmt='%d', padding=3)
ax.set_title("Q1 vs Q2 Regional Sales")
ax.set_xlabel("Region")
ax.set_ylabel("Sales (USD)")
ax.legend()
plt.show()
```

Using seaborn, create a scatter plot with study_hours on the x-axis and exam_scores on the y-axis. Add a title and axis labels.
