```python
import matplotlib.pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales  = [120, 145, 98, 167, 210]
```

```python
plt.plot(months, sales)
plt.show()
```

```python
plt.figure(figsize=(3,2))
plt.plot(months, sales, color='green', linewidth=3, marker='x')
plt.title("Monthly Sales Performance", fontsize=14, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.show()
```

```python
# comparison of different categories: bar plot
subjects = ["Math", "Science", "English", "History"]
scores   = [88, 92, 76, 85]
```

```python
plt.bar(subjects, scores)
plt.show()
```

# Seaborn

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
sns.histplot(data=df, x='total_bill')
```

```python
sns.histplot(data=df, x='tip', bins=10)
```

```python
sns.histplot(data=df, x='tip', binwidth=1)
```

```python
# relation between 2 numeric variables: scatterplot
sns.scatterplot(data=df, x='total_bill', y='tip')
```

```python
# hue-property
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex')
```

```python
# Boxplot
sns.boxplot(data=df, x='tip')
```

```python
sns.boxplot(data=df, y='tip')
```

```python
# numeric vs categorical

sns.boxplot(data = df, x='sex', y='tip')
```

```python
sns.boxplot(data = df, x='sex', y='total_bill')
```

```python
sns.lineplot(data=df, x='day', y='total_bill')
sns.barplot(data=df, x='day', y = 'total_bill')
plt.show()
```

```python
df
```

```python
sns.pairplot(data=df)
```

```python
# correlation analysis: [-1,1]

corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

```python
sns.heatmap(corr_matrix, cmap="Blues")
```
