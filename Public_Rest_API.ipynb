Send a GET request to the endpoint using the requests library inside a try-except block. Print the HTTP status code. If the request is successful (status code 200), print the total number of countries retrieved. If it fails, print a descriptive error message.

```python
import requests
import json
import pandas as pd
```

```python
def countries_data():
  """Countries data from restcountries.com"""
  url = "https://restcountries.com/v3.1/all"
  params = {"fields":"name,population,region,area"}

  try:
    response = requests.get(url, params, timeout=10)
    print(f"HTTP status code: {response.status_code}")
    if response.status_code == 200:
      print(f"Total number of countries retrieved: {len(response.json())}")

  except requests.exceptions.Timeout:
    print("Request timed out — server did not respond in 10 seconds.")
    return None
  except requests.exceptions.ConnectionError:
    print("Connection failed — check your internet connection.")
    return None
  except:
    print("An unexpected error occurred. Pls try again and search the detailed explaination on AI")
    return None

```

```python
countries_data()
```

Parse the JSON response and extract the following fields for each country: common_name, population, region, and area. Store the results in a Pandas DataFrame and display the first 10 rows.

```python
url = "https://restcountries.com/v3.1/all"
params = {"fields":"name,population,region,area"}
response = requests.get(url, params)
count_data = response.json()
df = pd.DataFrame(count_data)
df['common_name'] = df['name'].apply(lambda x: x['common'])
df.drop(columns = ['name'], inplace = True)
df = df[['common_name', 'population', 'region', 'area']]
df.head(10)
```

Which region has the highest total population?

```python
dm = df.groupby('region')['population'].sum()
dm = dm.reset_index()
dm[dm['population'].max() == dm['population']]['region']
```

```python
max_pupulation = df.groupby('region')['population'].sum().sort_values(ascending=False).reset_index()['region'][0]
print(f"The region with maximum population is {max_pupulation}.")
```

Which country has the largest area?

```python
max_area = df.groupby('common_name')['area'].sum().sort_values(ascending=False).reset_index()['common_name'][0]
print(f"The country with maximum area is {max_area}.")
```

```python
max_area = df['area'].max()

df[df['area'] == max_area]['common_name']
print(f"The country with maximum area is {max_area}.")
```
