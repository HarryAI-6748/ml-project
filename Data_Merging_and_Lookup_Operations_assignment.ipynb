=== TASK 3: Lookup and Automation ===

Lookup Operation:

Creating a dictionary mapping student_id to exam_score from the scores DataFrame

```python
stn_scr = dict(zip(stn_df['student_id'], scr_df['exam_score']))
print(f"The lookup dictionary is: {stn_scr}")
```

Using the .map() function to add exam scores to the students DataFrame

```python
stn_df['exam_score'] = stn_df['student_id'].map(stn_scr)
stn_df
```

Displaying students with their scores (showing NaN for students without scores)

```python
stn_df['exam_score']
```

Performance Comparison

Implementing the same score addition using pd.merge() with a left join

```python
pd.merge(stn_df, scr_df, how = 'left', on = 'student_id').drop(columns = ['exam_score_y'])
```

Why lookup (map) is more efficient than merge for this scenario?

Lookup is more efficient because the merge operations applied here is executed to entire arrays of data at once, rather than element by element. So it reduces the time and optimises memory required to complete this task.

Simple Automation: Defining the funtion

```python
def auto_merge(df1, df2, join_type, key_column):
  mrg_dict = pd.merge(df1, df2, how = join_type, on = key_column)
  formatted_output = {
      'result_df': mrg_dict,
      'row_count': mrg_dict.shape[0],
      'join_type': join_type,
  }
  fm_op = print("Rows in Result:", mrg_dict.shape[0], "\n", "Join type:", join_type, "\n", f"Result preview:\n {mrg_dict}")
  return fm_op
```
