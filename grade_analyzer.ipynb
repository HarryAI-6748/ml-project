```python
def process_scores(students):
    """This function takes a dictionary of student names and their marks, calculates the average score for each student, and returns a new dictionary with student names as keys and their average scores as values."""
    print("Hi, welcome to the Student grade analyser!", end="\n\n")
    student_avg = {}
    students_grades = {}
    while True:
        student_name = input("Pls enter your name (Enter 'quit' to stop): ")
        if student_name != "quit":
            mrk = int(input("Physics marks out of 100: "))
            if mrk > 100:
              print("Invalid Input")
              break
            mrk1 = int(input("Chemistry marks out of 100: "))
            if mrk1 > 100:
              print("Invalid Input")
              break
            mrk2 = int(input("Maths marks out of 100: "))
            if mrk2 > 100:
              print("Invalid Input")
              break
            print("Thanks. Pls enter next student details.", end="\n\n")
            marks = [mrk, mrk1, mrk2]
            avg_score = sum(marks) / len(marks)
            rounded_avg = round(avg_score, 2)
            student_avg[student_name] = rounded_avg
        else:
            break

    return(student_avg)


#
def classify_grades(student_avg):
    students_grades = {}
    for student_name, rounded_avg in student_avg.items():
        if rounded_avg >= 90:
            grade = "A"
        elif 89 >= rounded_avg >= 75:
            grade = "B"
        elif 74 >= rounded_avg >= 60:
            grade = "C"
        elif rounded_avg <= 59:
            grade = "F"
        students_grades[student_name] = (rounded_avg, grade)

    return students_grades


def generate_report(student_avg, students_grades):
    students_grades = classify_grades(student_avg)
    passing_threshhold = 70
    total_students = 0
    passed = 0
    failed = 0
    for student_name, (rounded_avg, grade) in students_grades.items():
      if rounded_avg >= passing_threshhold:
          status = "Pass"
      else:
          status = "Fail"
    for student_name, (rounded_avg, grade) in students_grades.items():
      total_students = len(student_avg)
      if rounded_avg >= passing_threshhold:
        passed += 1
      else:
        failed += 1
    print("===== Student Grade Analyser =====")
    for student_name, (rounded_avg, grade) in students_grades.items():
        print(f"{student_name.title()}     | Avg: {rounded_avg}  | Grade: {grade}  | Status: {status.upper()} ")
    print("================================")
    print(f"Total Students : {total_students}")
    print(f"Passed         : {passed}")
    print(f"Failed         : {failed}")

x = process_scores({})
y = classify_grades(x)
generate_report(x, y)
```
