Writing a Dictionary corresponding to the table in the question

```python
student = {
    "StudentID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Age": [20, 21, 19, 22, 20],
    "Marks": [88, 75, 92, 65, 78],
    "Grade": ["A", "B", "A", "C", "B"]
}
```

Importing pandas as pd and converting the dictionary into a data frame. Then adding that data frame into a file named student.csv

```python
import pandas as pd
st_df = pd.DataFrame(student)
st_df.to_csv('student.csv', index = False)
```

Reading the csv file using read_csv method

```python
pd.read_csv('student.csv')
```
