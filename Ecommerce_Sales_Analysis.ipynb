Ecommerce Sales Data

```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
ecc_data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

ecc_df = pd.DataFrame(ecc_data)
print(ecc_df.head(10))
print(f"\nDataset shape: {ecc_df.shape}")
```

Task 1: Basic Grouping and Single Aggregations

Calculating total sales amount for each region using groupby() and sum()

```python
ecc_df.groupby('region')['sales_amount'].sum().reset_index()
```

Counting the number of transactions for each product category using groupby() and count()

```python
ecc_df.groupby('category')['sales_amount'].count().reset_index()
```

Calculating the average sales amount per salesperson using groupby() and mean()

```python
ecc_df.groupby('salesperson')['sales_amount'].mean().round(2).reset_index()
```

Sorting the regional sales results in descending order using sort_values(ascending=False) to identify the top-performing region

```python
ecc_df.groupby('region')['sales_amount'].sum().sort_values(ascending = False).reset_index()
```

Task 2: Multi-Column Grouping and Multiple Aggregations

Grouping by both region AND category to calculate total sales for each combination using: df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()


```python
ecc_df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()
```

For each salesperson, calculating three metrics simultaneously using the agg() method:
    Total sales ('sum')
    Average sales ('mean')
    Number of transactions ('count')

```python
ecc_df.groupby('salesperson').agg({'sales_amount' : ['sum', 'mean'], 'transaction_id' : 'count'}).reset_index()# method 2
```

```python
ecc_df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()# method 1
```

Sorting the salesperson results by total sales in descending order to identify the top performer

```python
y = ecc_df.groupby('salesperson')['sales_amount'].agg(['sum', 'count'])
y.sort_values(by = 'sum', ascending = False).reset_index().head(1)
```

Using .idxmax() on the grouped category sales to find which category has the maximum total revenue

```python
ecc_df.groupby('category')['sales_amount'].sum().idxmax()
```

Custom Aggregation and Complete Sales Report

Create a comprehensive sales analysis report:


Define a custom aggregation function that calculates the sales range (max - min) for each group:

```python
def sales_range(x):
  return x.max() - x.min()
```

Applying this custom function along with standard aggregations to analyze sales by region:

```python
ecc_df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
```

Creating a final summary report that shows for each region:

    Total number of transactions (using customer_id with count)
    Total sales amount  
    Average transaction value

```python
ecc_df.groupby('region').agg({
    'transaction_id' : 'count',
    'sales_amount' : ['sum', 'mean']
}).round(2).reset_index()
```

The multilevel column structure takes input as key value pairs in which for each region, total counts of transactions, total sum of sales and the average of sales amount are displayed altogether which is possible because multiple aggregation functions are applied to one or more columns.

Single aggregations refer to applying a single function to one or more columns, or applying a list of functions without using a dictionary
